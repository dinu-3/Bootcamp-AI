```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\Owner\Downloads\ipl_matches_smalldata.csv")
print("Columns in dataset:", df.columns)
player_runs = df.groupby("batter")["batsman_runs"].sum().reset_index()
top_10_players = player_runs.sort_values(by="batsman_runs", ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_10_players["batter"], top_10_players["batsman_runs"], color='blue')
plt.xlabel("Players", fontsize=12)
plt.ylabel("Total Runs", fontsize=12)
plt.title("Top 10 Players by Runs in IPL", fontsize=14)
plt.xticks(rotation=45)  
plt.grid(axis='y', linestyle='--', alpha=0.7) 
plt.show()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
text = """Kabaddi is a popular contact sport that originated in ancient India and is now played internationally. The game is played between two teams, with each team taking turns to send a "raider" into the opponent's half. The raider must tag as many defenders as possible and return to their side while continuously chanting "kabaddi." If the raider is tackled and prevented from returning, they are out. The sport requires agility, strength, and strategic teamwork. Kabaddi has gained widespread recognition, with professional leagues such as the Pro Kabaddi League in India increasing its global appeal. It is now included in international tournaments, including the Asian Games."""
print(text)
words = word_tokenize(text.lower()) 
print(words)
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.isalpha() and word not in stop_words]
print(filtered_words)
word_freq = Counter(filtered_words)
print(word_freq)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off') 
plt.show()
```
